```python
import pandas as pd

dt = pd.read_json("/content/review.json.gz", lines=True)
dt.head()
```

```python
dt.info
```

```python
df = dt.mask(dt.eq('None')).dropna()
df['time'] = pd.to_datetime(df['time'])
df
```

```python
import plotly.express as px

gist_employee_residence = dt.groupby(['rating']).agg({"text": "count"})
gist_employee_residence.dropna(inplace=True)
fig = px.bar(gist_employee_residence, x = gist_employee_residence.index, y = gist_employee_residence['text'],
                                                                                                     title = 'Соотношение количества отзыва и рейтенга')
fig.show()
```
